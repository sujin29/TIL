### apply
어제 하던거 이어서 titanic으로 진행

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# 15세 이하는 child, 15~60까지는 adult, 61 이상은 elderly로 분류해서 
# age_division컬럼에 저장. lambda식으로 apply 함수를 이용해서 구현.

func = lambda x: 'child' if x <= 15 else ('adult' if x <=60 else 'elderly')
titanic['age_division'] =  titanic['age'].apply(func)
titanic.head()
```

- lambda 식에는 `elif`가 존재하지 않는다

elif가 많이 필요한 경우 : lambda에 def로 정의한 함수를 apply한다

```python
# 5살 이하는 baby, 12살 이하는 child, 19살 이하는 teen, 24살 이하는 student, 
# 39살 이하는 young adult, 60살 이하는 adult, 그 이상은 elderly라고 한다면

def get_category(age):
    category = ''
    if age <= 5:
        category = 'baby'
    elif age <= 12 :
        category = 'child'
    elif age <= 19:
        category = 'teen'
    elif age <= 24:
        category = 'student'
    elif age <= 39:
        category = 'young adult'
    elif age <= 60:
        category = 'adult'
    else:
        category = 'elderly'
    return category

titanic['category'] = titanic['age'].apply(lambda x : get_category(x))
titanic.head()
```

### columns이 아니라 DataFrame이 apply함수에 넘어올 수 있다

```python
# 20살 이하는 child 아닌 것은 adult로 해보기

func = lambda df: 'child' if df['age']<=20 else 'adult'
titanic['child/adult'] = titanic.apply(func, axis = 1)
titanic.head()
```

```python
# 승객에 대한 나이와 성별에 의한 카테고리를 'cat'으로 추가한다.
# 조건1 : 20살이 넘으면 성별을 그대로사용
# 조건2 : 20살미만이면 성별에 관계없이 'child'라고 정의
# null은 'i don't know'로 넣어보자

func1 = lambda f: f['sex'] if f['age']>=20 else 'child'
func2 = lambda f: "I don't know" if pd.isnull(f['age']) else func1(f)

titanic['cat'] = titanic.apply(func2, axis=1)
titanic.head()
titanic['cat'].value_counts()
```

### fillna()
- NaN값을 괄호 안의 값으로 변경한다
- 기존 데이터를 대체하지 않는다. 할당해야 한다.

```python
sample_df = pd.DataFrame({
'A':[1, 3, 4, 3, 4],
'B':[2, 3, 1, 2, 3],
'C':[1, 5, 2, 4, 4]
})
sample_df
```

```python
sample_df['A'].value_counts()
```

NaN값 만들기

```python
sample_df = sample_df.apply(pd.value_counts)
sample_df
```

NaN값을 0으로 채우기

```python
sample_df = sample_df.fillna(0)
sample_df
```

### 타이타닉 데이터로 실습

타이타닉 가져오기

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
```

타이타닉 승객 중 나이를 명시하지 않은 고객은 나이를 명시한 고객의 평균 나이값으로 대체.   
타이타닉 승객에 대해 나이와 성별에 의한 age_gender_cat 열을 만들고  
1. 성별을 나타내는 문자열 male 또는 female로 시작한다  
2. 성별을 나타내는 문자열 뒤에 나이를 나타내는 문자열로 변경  
3. 예시) 남성의 나이가 27이라면 -> male27  

```python
# 타이타닉 승객 중 나이를 명시하지 않은 고객은 
# 나이를 명시한 고객의 평균 나이값으로 대체.

titanic['age'] = titanic['age'].fillna(titanic['age'].mean())
titanic.count()
```

```python
# 타이타닉 승객에 대해 나이와 성별에 의한 age_gender_cat 열을 만들고
# 성별을 나타내는 문자열 male 또는 female로 시작한다
# 성별을 나타내는 문자열 뒤에 나이를 나타내는 문자열로 변경
# 예시) 남성의 나이가 27이라면 -> male27


titanic['age_gender_cat'] = titanic['sex'] + titanic['age'].astype(int).astype(str)
titanic.head()
```

## 데이터프레임 인덱스 조작 `set_index`, `reset_index`
- merge하기 위해 사용
- `set_index`: 기존 행 인덱스를 제거하고 데이터 열 중 하나를 인덱스로 세팅
- `reset_index`: 기존 행 인덱스를 데이터의 열로 추가. 0으로 시작하는 새 인덱스 생긴다

```python
np.random.seed(100)
index_df = pd.DataFrame(np.vstack([list('ABCDE'),
                                 np.round(np.random.rand(3, 5), 2)]).T,
                       columns=['col01', 'col02', 'col03', 'col04'])
index_df
```

```python
# index_df의 col1열을 인덱스로 세팅해서 index_df2에 할당
index_df2 = index_df.set_index('col01')
index_df2
```

```python
# index_df2의 col2열을 인덱스로 세팅해서 index_df3에 할당
index_df3 = index_df2.set_index('col02')
index_df3
```

```python
# index_df2의 인덱스를 reset
index_df2.reset_index()
```

```python
# index_df2의 인덱스를 reset, col1은 버리기
index_df2.reset_index(drop = True)
```

`drop=True`하면 기존 인덱스가 데이터의 컬럼으로 들어가지 않고 없어진다 

### 실습

5명의 학생의 국어, 영어, 수학 점수를 나타내는 데이터프레임을 만든다.


1. 
학생 이름을 나타내는 열을 포함시키지 않고  
데이터프레임 df_score1 을 생성한 후,   
df_score1.index 속성에 학생 이름을 나타내는 열을 지정하여 인덱스를 지정한다.  
reset_index 명령으로 이 인덱스 열을 명령으로   
일반 데이터열로 바꾸여 데이터프레임 df_score2을 만든다.  

2.
학생 이름을 나타내는 열이 일반 데이터 열을 포함하는 데이터프레임 df_score2에  
set_index 명령을 적용하여 다시 학생 이름을 나타내는 열을 인덱스로 변경한다.
 


```python
# 5명의 학생의 국어, 영어, 수학 점수를 나타내는 데이터프레임을 만든다.
np.random.seed(2)
df_score1 = pd.DataFrame(np.random.randint(100, size=(5,3)),
                        columns = ['국어', '영어', '수학'])
df_score1
```

```python
# df_score1.index 속성에 학생 이름을 나타내는 열을 지정하여 인덱스를 지정한다.
# reset_index 명령으로 이 인덱스 열을 명령으로
# 일반 데이터열로 바꾸여 데이터프레임 df_score2을 만든다.

name = ['짱구', '철수', '유리', '훈이', '맹구']
df_score1['name'] = name
df_score1 = df_score1.set_index('name')
df_score1

df_score2 = df_score1.reset_index()
df_score2
```

```python
# 학생 이름을 나타내는 열이 일반 데이터 열을 포함하는 데이터프레임 df_score2에
# set_index 명령을 적용하여 다시 학생 이름을 나타내는 열을 인덱스로 변경한다.

df_score2.set_index('name')
```

### DataFrame merge

```python
data1 = {
    '학번':[1, 2, 3, 4],
    '이름':['섭섭해', '김한준', '김선림', '최호진'],
    '학년':[2, 4, 1, 3]
}

data2 = {
    '학번':[1, 2, 4, 5],
    '학과':['CS', 'MATH', 'CS', 'MATH'],
    '학점':[2.4, 4.5, 3.4, 4.2]
}
```

```python
stu_df = pd.DataFrame(data1)
major_df = pd.DataFrame(data2)
display(stu_df)
display(major_df)
```

```python
pd.merge(stu_df, major_df)
```

pd.merge 디폴트는 `on = 공통컬럼`, `how = inner`(매칭되지 않는 것은 누락)

```python
pd.merge(stu_df, major_df, on ='학번', how='inner')
```

left - left 데이터는 모두 출력. 비어있는 값은 NaN으로 출력

```python
pd.merge(stu_df, major_df, how='left')
```

right - right데이터는 모두 출력. 비어있는 값은 NaN으로 출력

```python
pd.merge(stu_df, major_df, how='right')
```

outer - 모든 데이터 출력. 비어있는 값은 NaN으로 출력

```python
pd.merge(stu_df, major_df, how='outer')
```

inner는 교집합만 출력  
left는 왼쪽꺼는 모두 출력  
right는 오른쪽꺼는 모두 출력  
outer는 전부 다 출력

### 컬럼 인덱스가 다른 경우 (동일한 컬럼이 없는 경우)

```python
data1 = {
    '학번':[1, 2, 3, 4],
    '이름':['섭섭해', '김한준', '김선림', '최호진'],
    '학년':[2, 4, 1, 3]
}

data2 = {
    '과목코드':[1, 2, 4, 5],
    '학과':['CS', 'MATH', 'CS', 'MATH'],
    '학점':[2.4, 4.5, 3.4, 4.2]
}
```

```python
stu_df = pd.DataFrame(data1)
major_df = pd.DataFrame(data2)
display(stu_df)
display(major_df)
```

```python
pd.merge(stu_df, major_df, left_on = '학번', right_on = '과목코드', how='outer')
```

동일한 컬럼이 없는 경우, 어떤 컬럼을 기준으로 merge할지 알려줘야 한다  
왼쪽 컬럼에서는 '학번', 오른쪽 컬럼의 경우 '과목코드'를 기준으로 merge한다  
how='inner'디폴트

같은 컬럼이 두 개 있을 때 (품종, 꽃잎너비)

```python
iris_df01 = pd.DataFrame({
    '품종': ['setosa', 'setosa', 'virginica', 'virginica'],
    '꽃잎너비': [1.4, 1.3, 1.5, 1.3],
    '개화시기': ['202012', '202010', '202009', '202010']
})

iris_df02 = pd.DataFrame({
    '품종': ['setosa', 'virginica', 'virginica', 'versicolor'],
    '꽃잎너비': [0.4, 0.3, 0.5, 0.3]
})
```

```python
display(iris_df01)
display(iris_df02)
```

### columns인덱스가 아닌 row인덱스를 기준으로 병합

```python
pop_df01 = pd.DataFrame({
    'city':['seoul', 'seoul', 'seoul', 'busan', 'busan'],
    'year':[2010, 2005, 2020, 2018, 2015],
    'pop':[1234567, 2345678, 3456789, 4567890, 5678901]
})
pop_df01
```

```python
pop_df02 = pd.DataFrame(np.arange(12).reshape((6,2)),
            index = [['busan' , 'busan', 'seoul' , 'seoul' , 'seoul' , 'seoul'], 
                     [2010, 2005, 2020, 2018, 2015, 2010]],
            columns = ['col01', 'col02'])
pop_df02
```

```python
pd.merge(pop_df01, pop_df02, right_index=True, left_on = ['city', 'year'])
```

right데이터에서는 index를 사용하고, left데이터에서는 columns를 사용한다.  
right데이터의 인덱스의 depth가 2니까 left데이터에서도 2개의 columns를 이용해야 한다.  
이것도 디폴트는 inner

### 단일 인덱스에서 index를 기준으로 merge

```python
data1 = { "이름" : ["이지안","박동훈","이순신","강감찬"],
          "학년" : [2,4,1,3]}

data2 = { "학과" : ["CS","MATH","MATH","CS"],
          "학점" : [3.4,2.9,4.5,1.2]}
```

```python
df01 = pd.DataFrame(data1, index = [1, 2, 3, 4])
df02 = pd.DataFrame(data2, index = [1, 2, 4, 5])

display(df01)
display(df02)
```

```python
pd.merge(df01, df02, left_index = True, right_index = True)
```

index를 기준으로 merge한다.  
디폴트는 inner 조인. 겹치는 1, 2, 4만 가져온다
