파일 불러오기

```python
files = glob('../../data/oil_data/지역*xls')
files
```

```python
tmp_raw = []

for file in files:
    tmp = pd.read_excel(file, header = 2)
    tmp_raw.append(tmp)
tmp_raw
```

```python
oil_station = pd.concat(tmp_raw)
oil_station
```

```python
oil_station.info()
```

feature selection

```python
oil_station.head()
```

```python
stations = pd.DataFrame({
    'oil_station' : oil_station['상호'],
    'address': oil_station['주소'],
    'price': oil_station['휘발유'],
    'self_yn': oil_station['셀프여부'],
    'company': oil_station['상표']
})

stations.head()
```

```python
# 파생 변수(구) 추가

stations['district'] = [addr.split()[1] for addr in stations['address']]
stations['district']
```

```python
stations['district'].unique()
```

```python
# 서울특별시, 특별시 대체

stations.loc[stations['district'] == '서울특별시', 'district'] = '성동구'
stations.loc[stations['district'] == '특별시', 'district'] = '도봉구'

stations['district'].unique()
```

```python
# 가격정보 확인 가격정보가 기입되지 않은 주유소는 제거

stations = stations[stations['price'] != '-']
```

```python
# price 컬럼 타입을 숫자로 변경

stations['price'] = stations['price'].astype(int)
```

```python
stations.info()
```

```python
# index 처리

stations.reset_index(inplace = True)
stations.tail()
```

```python
# 필요없는 index 제거

stations = stations.drop('index', axis = 1)
```

```python
stations.tail()
```

```python
# self_yn분포 확인

sns.boxplot(data=stations, x = 'self_yn', y = 'price')
```

```python
# 회사 별 가격

plt.figure(figsize=(20, 10))

sns.boxplot(data = stations, x = 'company', y = 'price', palette = 'Set3')
sns.swarmplot(data = stations, x = 'company', y = 'price', palette = 'Set3')

plt.show()
```

구 별 가격정보

```python
import folium
import json
import googlemaps
```

```python
# 피봇 이용해서 구 별 가격정보로 변경하고 가격

district_pivot = stations.pivot_table(values = 'price', 
                                      index = 'district', aggfunc = 'mean')
district_pivot.head()
```

```python
geo_str = json.load(open('../../data/02. skorea_municipalities_geo_simple.json', 
                        encoding = 'utf-8'))
```

```python
geo_str
```

```python
seoul_map = folium.Map([37.566023, 126.978203],
                      zoom_start = 10.5)
folium.Choropleth(geo_data = geo_str,
                 data = district_pivot,
                 columns = [district_pivot.index, 'price'],
                 key_on = 'feature.id').add_to(seoul_map)

seoul_map
```

```python
# 주유가격 상위 10, 하위 10을 map에 시각화

oil_station_top10 = stations.sort_values(by='price', ascending = False).head(10)
oil_station_top10

oil_station_bottom10 = stations.sort_values(by='price', ascending = True).head(10)
oil_station_bottom10
```

```python
gmap_key = ' '
gmaps = googlemaps.Client(key = gmap_key)
```

```python
lat = []
lng = []

for n in oil_station_top10.index:
    tmp_addr = str(oil_station_top10['address'][n])
#     print(tmp_addr)
    tmp_map = gmaps.geocode(tmp_addr)
#     print(tmp_map)
    tmp_loc = tmp_map[0].get('geometry')
    lat.append(tmp_loc['location']['lat'])
    lng.append(tmp_loc['location']['lng'])
    
# print(lat)
# print(lng)

oil_station_top10['lat'] = lat
oil_station_top10['lng'] = lng

oil_station_top10.head()
```

```python
lat = []
lng = []

for n in oil_station_bottom10.index:
    tmp_addr = str(oil_station_bottom10['address'][n])
#     print(tmp_addr)
    tmp_map = gmaps.geocode(tmp_addr)
#     print(tmp_map)
    tmp_loc = tmp_map[0].get('geometry')
#     print(tmp_loc)
    lat.append(tmp_loc['location']['lat'])
    lng.append(tmp_loc['location']['lng'])
    
oil_station_bottom10['lat'] = lat
oil_station_bottom10['lng'] = lng
```

```python
oil_station_bottom10.head()
```

```python
for name, lat, lng in zip(oil_station_top10['oil_station'], oil_station_top10['lat'], oil_station_top10['lng']):
    folium.Marker([lat, lng],
                 popup = name,
                 icon = folium.Icon(color = 'red')).add_to(seoul_map)
seoul_map
```

```python
for name, lat, lng in zip(oil_station_bottom10['oil_station'], oil_station_bottom10['lat'], oil_station_bottom10['lng']):
    folium.Marker([lat, lng],
                 popup = name,
                 icon = folium.Icon(color = 'blue')).add_to(seoul_map)
    
seoul_map
```
