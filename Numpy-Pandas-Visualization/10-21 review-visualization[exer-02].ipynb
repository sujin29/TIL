지역별 연령층 분포
1. 컬럼 추출
2. 컬럼 조작 (지역명)
3. groupby -> pivot
4. 시각화

```python
data.head()
```

```python
region_gen_df = data[['태어난연도', '지역구분']]
region_gen_df
```

```python
region_gen_df['나이'] = dt.datetime.now().year - region_gen_df['태어난연도'] +1
region_gen_df.head()
```

```python
region_gen_df['연령대'] = region_gen_df['나이'] //10 *10
region_gen_df.head()
```

```python
region_code = pd.DataFrame({
    'code':list(range(1,8)),
    '지역명':['서울', '수도권(인천/경기)', '부산/경남/울산', '대구/경북',
          '대전/충남', '강원/충북', '광주/전남/전북/제주도']
})
region_code
```

```python
region_gen_merge_df = pd.merge(region_gen_df, region_code, left_on = '지역구분', right_on ='code')
region_gen_merge_df.head()
```

```python
region_gen_merge_df = region_gen_merge_df[['나이', '연령대', '지역명']]
region_gen_merge_df.head()
```

```python
# 연령층은 나이 구간을 설정 
# 나이 <30: 청년층,  나이 < 60: 중년층, 그렇지 않으면 노년층
# 연령층 파생컬럼 추가

def age(x):
    g=''
    if x<30:
        g = '청년층'
    elif x<60:
        g = '중년층'
    else:
        g = '노년층'
    return(g)

region_gen_merge_df['연령층'] = region_gen_merge_df['나이'].apply(lambda x: age(x))
region_gen_merge_df.tail()
```

### numpy.select
numpy.select(condlist, choicelist, default=0)[source]
- np.select(조건리스트, 조건에 따른 선택리스트)

```python
condition = [region_gen_merge_df['나이'] <30,
             region_gen_merge_df['나이'] <60,
             region_gen_merge_df['나이'] >=60 ]
choice = ['청년층', '중년층', '노년층']
region_gen_merge_df['연령층2'] = np.select(condition, choice)
region_gen_merge_df
```

```python
# 지역과 연령층에 따른 빈도분포 확인
region_gen_count = region_gen_merge_df.groupby(['지역명', '연령층'], as_index = False)[['나이']].count()
region_gen_count
```

```python
region_gen_count.rename(columns = {'나이':'명'}, inplace = True)
region_gen_count
```

```python
region_gen_pivot = region_gen_count.pivot('지역명', '연령층', '명')
region_gen_pivot
```

```python

region_gen_pivot.plot.bar(rot = 0)
plt.title('지역별 연령층 분포')
plt.xlabel('지역')
plt.ylabel('연령층')
plt.show()
```

```python
# 연령대 별로 나눠서 지역별 pie chart
region_gen_pivot
```

```python
plt.rcParams['figure.figsize'] = (20, 10)

fig = plt.figure()
area01 = fig.add_subplot(1, 3, 1)
area02 = fig.add_subplot(1, 3, 2)
area03 = fig.add_subplot(1, 3, 3)

region_gen_pivot['청년층'].plot.pie(ax = area01, autopct = '%0.1f%%')
area01.title.set_text('청년층')
area01.set(ylabel = None)

region_gen_pivot['중년층'].plot.pie(ax = area02, autopct = '%0.1f%%')
area02.title.set_text('중년층')
area02.set(ylabel = None)

region_gen_pivot['노년층'].plot.pie(ax = area03, autopct = '%0.1f%%')
area03.title.set_text('노년층')
area03.set(ylabel = None)


plt.show()
```
