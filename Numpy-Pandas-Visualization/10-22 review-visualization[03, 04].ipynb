```python
# for loop돌려서 maker에 이름 추가 - 방법 1

g_map = g.Map(location=[37.566644, 126.978424],
              zoom_start = 10)

for idx in range(len(seoul_univ)):
    mark = g.Marker(seoul_univ.loc[idx, ['위도', '경도']],
                    popup = seoul_univ.iloc[idx, 0],
                    icon = g.Icon(color = 'blue'))
    mark.add_to(g_map)
g_map
```

```python
# for loop돌려서 maker에 이름 추가 - 방법 2

g_map = g.Map(location=[37.566644, 126.978424],
              zoom_start = 10)

for name, lat, lng in zip(seoul_univ.index,seoul_univ.위도, seoul_univ.경도):
    g.Marker([lat, lng], 
             popup = name, 
             icon = g.Icon(color = 'gray')).add_to(g_map)
g_map
```

```python
# 마커에 옵션 넣기


g_map = g.Map(location=[37.566644, 126.978424],
              zoom_start = 10)

for name, lat, lng in zip(seoul_univ.name,seoul_univ.위도, seoul_univ.경도):
    g.CircleMarker([lat, lng],
                   popup = name,
                   icon = g.Icon(),
                   color = 'coral',
                   fill_color = 'coral',
                   fill_opacity = .3).add_to(g_map)
g_map
```

`.values`

## barplot()
: 카테고리 값에 따른 실수 값의 평균과 편차를 표시하는 차트
- 평균: 막대의 높이
- 편차: 에러바

```python
# titanic으로 3개 그래프 그리기 (성별에 따른 생존율)

fig = plt.figure(figsize = (15, 5))
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

sns.barplot(x = 'sex', y = 'survived', data = titanic, ax = ax1)
sns.barplot(x = 'sex', y = 'survived', hue = 'pclass', data = titanic, ax = ax2)
sns.barplot(x = 'sex', y = 'survived', hue = 'pclass', dodge = False, data = titanic, ax = ax3)


plt.show()
```

```python
tips.head()
```

```python
# tips 요일 별 팁 평균

sns.barplot(x = 'day', y = 'tip', data = tips)

plt.show()
```

## countplot()
카테고리 값 별로 데이터 빈도를 표시하는 차트
- x축의 빈도를 세는 것이기 때문에 y축은 들어가지 않는다.

```python
# 클래스 별 countplot 그리기

sns.set_style('whitegrid')

fig = plt.figure(figsize = (15, 5))
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

sns.countplot(x = 'class', data = titanic, ax = ax1)
sns.countplot(x = 'class', hue = 'sex', data = titanic, ax = ax2)
sns.countplot(x = 'class', hue = 'sex', dodge = False, data = titanic, ax = ax3)


plt.show()
```

```python
# tips 요일 별 전체 팁 (빈도)

sns.countplot(x = 'day', data = tips, palette = 'Set3')
```

## Stripplot()
scatterplot처럼 모든 데이터를 점으로 시각화인데 범주형 데이터를 scatter
- `jitter = True` 설정하면 위치를 무작위로 바꿔서 겹치지 않게 해준다

```python
# 요일 별 전체 팁 분포

fig = plt.figure(figsize = (15, 5))
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)


sns.stripplot(x = 'day', y = 'tip', data = tips, jitter = False, ax = ax1)
sns.stripplot(x = 'day', y = 'tip', data = tips, jitter = True, ax = ax2)
sns.swarmplot(x = 'day', y = 'tip', data = tips, ax = ax3)

plt.show()
```

- stripplot에서 `jitter=False` 한 줄, `jitter=True` 옆으로 분포  
- swarmplot 절대 안겹침
- 오른쪽으로 갈 수록 분포 알기 더 쉽다

## swarmplot()
stripplot이랑 비슷한데 겹치는 부분 없이 시각화(분포 보기 더 좋다)

```python
# 요일 별 전체 팁 분포
sns.swarmplot(x = 'day', y = 'tip', data = tips)
```

```python
# 타이타닉 데이터를 이용하여
# 서브플롯을 2개 영역으로 만들고
# class 별 나이의 분포를 확인하라(stripplot(), swarmplot())

fig = plt.figure(figsize = (20, 10))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

sns.stripplot(x = 'class', y = 'age', data = titanic, ax = ax1)
sns.swarmplot(x = 'class', y = 'age', data = titanic, ax = ax2)

plt.show()
```

```python
# tips에서 요일별, 성별 전체 팁을 stripplot() 시각화

sns.stripplot(x = 'day', y = 'tip', hue = 'sex', palette='Set2', data = tips)
plt.show()
```

## boxplot, violin plot

```python
fig = plt.figure(figsize = (20, 10))

ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

sns.boxplot(x = 'alive', y = 'age', data = titanic, ax = ax1)
sns.boxplot(x = 'alive', y = 'age', hue = 'sex', data = titanic, ax = ax2)
sns.violinplot(x = 'alive', y = 'age', data = titanic, ax = ax3)
sns.violinplot(x = 'alive', y = 'age', hue = 'sex', data = titanic, ax = ax4)

plt.show()
```

## heatmap()
하나 이상의 카테고리 값의 변화를 색깔 변화로 시각화한 차트

```python
titanic.head()
```

```python
table = titanic.pivot_table(index = 'sex', columns = 'class', aggfunc = 'size')
table
```

```python
sns.heatmap(table, annot = True, fmt = 'd')
```

```python
# flights 연도, 월 별 승객 수에 대한 heatmap그려보기 (x축 연도, y축 월)

flights.head()
```

```python
f_table = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers', aggfunc = 'sum')
f_table
```

```python
sns.heatmap(f_table,
            annot = True, 
            fmt = 'd')

plt.show()
```
