### bar plot

```python
Y = [2, 3, 1]
X = np.arange(len(Y))

# x는 pclass
# y는 survived

xlabel = ['1등급', '2등급', '3등급']
plt.bar(X, Y)
plt.title('선실등급 별 생존자 수')
plt.xlabel=['선실등급']
plt.ylabel=['생존자 수']
plt.xticks(X, xlabel)
plt.yticks(Y)
```

```python
# horizontal

Y = [2, 3, 1]
X = np.arange(len(Y))

# x는 pclass
# y는 survived

xlabel = ['1등급', '2등급', '3등급']
plt.barh(X, Y)
plt.title('선실등급 별 생존자 수')
plt.xlabel=['선실등급']
plt.ylabel=['생존자 수']
plt.yticks(X, xlabel)
plt.xticks(Y)
plt.show()
```

### 양적자료

- boxplot으로 시각화
- 배기량(displ), 생산년도(year), 실린더개수(cyl), 도시연비(cty), 고속도로연비(hwy)

### 질적자료
- describe로 요약되지 않는 자료. 범주형, 순서형태.
- 도수분포표와 히스토그램으로 데이터의분포를 확인
- 제조회사, 모델명, 변속기종류, 구동방식, 연료종류, 자동차종류
- `value_counts()`로 데이터 빈도 확인

```python
# 문제 1) 자동차 배기량(displ)에 따라 고속도로 연비(hwy)가 다른지를 알아보자 
# - 배기량이 4이하인 자동차와 5이상인 자동차 중 
# 어떤 자동차의 고속도로 연비가 평균적으로 높은지 알아보자 (query)

print('4 이하', data_df[data_df['displ'] <= 4]['hwy'].mean())
print('5 이상', data_df[data_df['displ'] >= 5]['hwy'].mean())

print('query 4 이하: ', data_df.query('displ <= 4')['hwy'].mean())
print('query 5 이상: ', data_df.query('displ >= 5')['hwy'].mean())
```

```python
# 문제 2) 자동차 제조 회사에 따라 도시 연비가 다른지 알아보려고 한다. 
# audi와 toyota 중 어느 manufacturer의 cty가 평균적으로 더 높은지 알아보자

manu_data = pd.DataFrame(data_df.groupby('manufacturer')['cty'].mean())
manu_data.loc[['audi', 'toyota']]
```

```python
# 문제 3) chevolet, ford, honda 자동차의 고속도로 연비 평균을 알아보려 한다. 
# 이 회사들의 데이터를 추출한 후 hwy 전체 평균을 확인하시오

manu3 = ['chevrolet', 'ford', 'honda']
new_data = data_df.query('manufacturer in @manu3')
new_data['hwy'].mean()
```

query에 '외부 변수명' 또는 '함수'를 쓸 경우에는 이름 앞에 @를 붙여준다

```python
fl_price_df = pd.DataFrame({
    'c':2.35,
    'd':2.38,
    'e':2.11,
    'p':2.76,
    'r':2.22
}, index = ['price']).T
fl_price_df = fl_price_df.reset_index()
fl_price_df.columns = ['fl', 'price']
fl_price_df
```

```python
fl_df = pd.DataFrame({
    'fl':['c', 'd', 'e', 'p', 'r'],
    'price':[2.35, 2.38, 2.11, 2.76, 2.22]
})
fl_df
```

```python
fl_df.info()
```

```python
fl_price_df.info()
```

```python
data_fl_merge_df = pd.merge(data_df, fl_df, on = 'fl')
data_fl_merge_df.head()
```

```python
data_fl_merge_df.loc[65, 'hwy'] = None # np.nan
data_fl_merge_df.loc[120, 'hwy'] = None
data_fl_merge_df.loc[154, 'hwy'] = None
data_fl_merge_df.loc[189, 'hwy'] = None
data_fl_merge_df.loc[219, 'hwy'] = None
data_fl_merge_df.loc[230, 'hwy'] = None
```

```python
# 문제 5) hwy변수의 결측값을 제외하고, 
# 구동방식 별 고속도로평균 연비를 알아보자

data_fl_merge_df.dropna(inplace=True)
drv_df = data_fl_merge_df.groupby('drv')['hwy'].mean()
drv_df
```

```python
# 문제 6) 구동방식 별 연비평균을 비교하기 위한 막대그래프

drv_df.plot.bar(rot = 0)
# plt.xlabel('drvvv')
plt.show()
```

```python
# 문제 7) 위에서 했던 방식과 동일하게 
# 구동방식 별 고속도로, 도시연비의 평균을 구해보고 
# 이를 데이터 프레임으로 만들어서 막대바로 시각화해보자

hwy_cty = data_fl_merge_df.groupby('drv')[['hwy', 'cty']].mean()
```

```python
hwy_cty.plot.bar(rot = 0)
```

```python
# 문제 8) 자동차(class)중에서 어떤 자동차가 가장 많은지 알아보려고 한다. 
# 종류별 빈도를 바플롯으로 시각화해보자

class_df = data_df['class'].value_counts()
class_df.plot.bar()
plt.show()
```

```python
# 문제 9) 어떤 회사에서 생산한 suv 차종의 도시 연비가 높은지를 알아보려고 한다. 
# suv 차종을 대상으로 평균 도시연비가 가장 높은 회사 다섯 곳을 
# 막대 바로 시각화해보자

suv_df = data_df[data_df['class']=='suv']
suv_cty_mean = pd.DataFrame(suv_df.groupby('manufacturer')['cty'].mean())
suv_cty_mean.sort_values(by='cty', ascending = False, inplace = True)
suv_cty_mean5 = suv_cty_mean.head()
```
